```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Abrimos nuestro CSV

```python
kickstarter = pd.read_csv("./Kickstarter_projects_Feb19.csv")

kickstarter.head(5)
```

### Creando una columna con la diferencia entre lo recaudado y el objetivo, se podría llegar a otro analisis de porque muchos proyectos no han funcionado... Pero ahora vamos a entrarnos en el global de los proyectos

```python
# Estos serían los proyectos que no han funcionado.

kickstarter['money']=kickstarter['usd_pledged']-kickstarter['goal_usd']
kickstarter[(kickstarter['money'] < 0) & (kickstarter['status'] != "successful")].sort_values(by='money')[0:15]
```

### Con el boxplot podemos ver dondé se reparte la cantidad recaudada por la mayoría de los proyectos

```python
kickstarter.boxplot(column="usd_pledged")
```

### Queremos saber cuales son los proyectos más frecuentes dentro de la web.

```python
kickstarter_category = kickstarter['main_category'].value_counts()

kickstarter_category.plot(kind = 'bar');
```

### Y también los que más dinero se marcan como objetivo y los que más dinero consiguen al llegar a la deadline.

```python
kickstarter_types= kickstarter.groupby(by=["main_category"]).agg({"goal_usd":['sum'],"usd_pledged":["sum"]})
kickstarter_types.plot.bar();
```

### También vamos a ver si existe correlación entre la duración que un proyecto está publico en la web y la cantidad de dinero que recauda

```python
kickstarter['usd_pledged'].corr(kickstarter['duration'])
```

```python
kickstarter.corr()
```

```python
kickstarter.corr(method = 'spearman')
```

### En este gráfico de dispersión podemos observar cómo se reparte el dinero recaudado según los días que el proyecto está publicado en la web.

```python
plt.scatter(kickstarter.duration,kickstarter.usd_pledged, alpha=0.5);
```

### Sacamos el TOP 10 de los paises con mayor cantidad de proyectos.

```python
kickstarter_country = kickstarter['country'].value_counts()[0:10]

kickstarter_country.plot(kind = 'bar');
```

### Nos interesa saber en que meses se lanzan más proyectos y en cuales terminan más.

```python
kickstarter_start = kickstarter.groupby(by='start_month').agg({'id':['count']})

kickstarter_start.plot(kind = 'line', style='.-')
plt.legend(loc='best', bbox_to_anchor=(1.0, 0.5));
```

```python
kickstarter_end = kickstarter.groupby(by='end_month').agg({'id':['count']})

kickstarter_end.plot(kind = 'line', style='.-')
plt.legend(loc='best', bbox_to_anchor=(1.0, 0.5));
```
